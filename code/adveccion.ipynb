$$\frac{\partial u}{\partial t} + a \frac{\partial u}{\partial x} = 0$$

```python
import numpy                       #here we load numpy
from matplotlib import pyplot      #here we load matplotlib
%matplotlib inline
```

```python
# Euler adelante

nx = 201 
dx = 2 / (nx-1)
nt = 100    #nt is the number of timesteps we want to calculate
dt = .005  #dt is the amount of time each timestep covers (delta t)
a = 1      #assume wavespeed of a = 1
print('dx = ', dx, '; dt = ', dt, '; ak/h = ', a*dt/dx)

u = numpy.ones(nx)      #numpy function ones()
u[int(.1 / dx):int(.6 / dx + 1)] = 2  #setting u = 2 between 0.5 and 1 as per our I.C.s

pyplot.plot(numpy.linspace(0, 2, nx), u, 'k')

un = numpy.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1, nx-1):
        u[i] = un[i] - a * dt / (2*dx) * (un[i+1] - un[i-1])

pyplot.plot(numpy.linspace(0, 2, nx), u)
```

```python
# Lax-Friedrichs

nx = 201 
dx = 2 / (nx-1)
nt = 100    #nt is the number of timesteps we want to calculate
dt = .005  #dt is the amount of time each timestep covers (delta t)
a = 1      #assume wavespeed of a = 1
print('dx = ', dx, '; dt = ', dt, '; ak/h = ', a*dt/dx)

u = numpy.ones(nx)      #numpy function ones()
u[int(.1 / dx):int(.6 / dx + 1)] = 2  #setting u = 2 between 0.5 and 1 as per our I.C.s

pyplot.plot(numpy.linspace(0, 2, nx), u, 'k')

un = numpy.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1, nx-1):
        u[i] = 0.5*(un[i-1]+un[i+1]) - a * dt / (2*dx) * (un[i+1] - un[i-1])

pyplot.plot(numpy.linspace(0, 2, nx), u)
```

```python
# Lax-Wendroff

nx = 2001 
dx = 2 / (nx-1)
nt = 1000    #nt is the number of timesteps we want to calculate
dt = .001  #dt is the amount of time each timestep covers (delta t)
a = 1      #assume wavespeed of a = 1
print('dx = ', dx, '; dt = ', dt, '; ak/h = ', a*dt/dx)

u = numpy.ones(nx)      #numpy function ones()
u[int(.1 / dx):int(.6 / dx + 1)] = 2  #setting u = 2 between 0.5 and 1 as per our I.C.s

pyplot.plot(numpy.linspace(0, 2, nx), u, 'k')

un = numpy.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1, nx-1):
        u[i] = un[i] - a*dt/(2*dx)*(un[i+1] - un[i-1]) + (a**2*dt**2)/(2*dx**2)*(un[i-1]-2*un[i]+un[i+1])

pyplot.plot(numpy.linspace(0, 2, nx), u)
```

```python
# Upwind
nx = 201 
dx = 2 / (nx-1)
nt = 100    #nt is the number of timesteps we want to calculate
dt = .005  #dt is the amount of time each timestep covers (delta t)
a = 1      #assume wavespeed of a = 1
print('dx = ', dx, '; dt = ', dt, '; ak/h = ', a*dt/dx)

u = numpy.ones(nx)      #numpy function ones()
u[int(.1 / dx):int(.6 / dx + 1)] = 2  #setting u = 2 between 0.5 and 1 as per our I.C.s

pyplot.plot(numpy.linspace(0, 2, nx), u, 'k')

un = numpy.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - a * dt / dx * (un[i] - un[i-1])

pyplot.plot(numpy.linspace(0, 2, nx), u)
```

```python
# Beam-Warming

nx = 21 
dx = 2 / (nx-1)
nt = 10    #nt is the number of timesteps we want to calculate
dt = .1  #dt is the amount of time each timestep covers (delta t)
a = 1      #assume wavespeed of a = 1
print('dx = ', dx, '; dt = ', dt, '; ak/h = ', a*dt/dx)

u = numpy.ones(nx)      #numpy function ones()
u[int(.1 / dx):int(.6 / dx + 1)] = 2  #setting u = 2 between 0.5 and 1 as per our I.C.s

pyplot.plot(numpy.linspace(0, 2, nx), u, 'k')

un = numpy.ones(nx)
for n in range(nt):
    un = u.copy()
    for i in range(2, nx-1):
        u[i] = un[i] - a*dt/(2*dx)*(3*un[i]-4*un[i-1] + un[i-2]) + (a**2*dt**2)/(2*dx**2)*(un[i]-2*un[i-1] + un[i-2])

pyplot.plot(numpy.linspace(0, 2, nx), u)
```

```python
# Leapfrog

nx = 21 
dx = 2 / (nx-1)
nt = 10    #nt is the number of timesteps we want to calculate
dt = .05  #dt is the amount of time each timestep covers (delta t)
a = 1      #assume wavespeed of a = 1
print('dx = ', dx, '; dt = ', dt, '; ak/h = ', a*dt/dx)

u = numpy.ones(nx)      #numpy function ones()
u[int(.1 / dx):int(.6 / dx + 1)] = 2  #setting u = 2 between 0.5 and 1 as per our I.C.s

pyplot.plot(numpy.linspace(0, 2, nx), u, 'k')

un = numpy.ones(nx)
unn = numpy.ones(nx)
for n in range(nt):
    if n == 0:
        unn = u.copy()
    else:
        unn = un.copy()
    un = u.copy()
    for i in range(1, nx-1):
        u[i] = unn[i] - a*dt/dx * (un[i+1] - un[i-1])

pyplot.plot(numpy.linspace(0, 2, nx), u)
```
